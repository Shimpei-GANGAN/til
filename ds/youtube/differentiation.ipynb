## グラフの練習

```python
import numpy as np
import matplotlib.pyplot as plt

# 二次関数を描画する
## 始点から終点までを等分に分割（デフォルトでは 50 等分）
x = np.linspace(-3, 3)
y = x**2 - 1
## dashed: 点線で描画
plt.plot(x, y, linestyle="dashed", label="x^2-1") 

y = x**3 /3 - x -1
plt.plot(x, y, linestyle="dashed", label="x^3/3-x-1") 


# グラフの整形
## 軸にタイトルをつける
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
## グリッド（縦線と横線）を入れる
plt.grid()
## 凡例の表示
plt.legend()

plt.show()
```

## 接線の表現
以下に`x=a`における、`f(x)`の微分の値を計算する式を示します。

$$
f ^{'}(a) = \lim _ {x \to a}\frac{f(x+a)-f(x)}{a}
$$

$x=a$における、$f(x)$の接線とは、

> 傾きが $f'(a)$で$x=a$, $y=f(a)$ を通る直線

のことであり、上の微分を用いて

$$
y = f ^{'}(a)(x-a) + f(a)
$$

のように定義される

```python
import numpy as np
import matplotlib.pyplot as plt

def quad_func(x):  # y=x^2-1を表す関数
    y = x**2 - 1
    return y

def tangent(func, x, a):
    delta_x = 1e-4
    derivative = (func(a+delta_x) - func(a)) / delta_x
    return derivative * (x - a) + func(a)

x = np.linspace(-3, 3)
y = quad_func(x)
plt.plot(x, y, label="y", linestyle="dashed")

## x = 2 の時の接線
a = 2
tangent = tangent(quad_func, x, a)
plt.plot(x, tangent, label="tangent")

## x = 2 の点も描画しておく
plt.scatter(a, quad_func(a))

plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.xlim([-3, 3])
plt.ylim([-3, 10])


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def quad_func(x):  # y=x^2+1を表す関数
    y = x**2 - 1
    return y

def tangent(func, x, a):
    delta_x = 1e-4
    derivative = (func(a+delta_x) - func(a)) / delta_x
    return derivative * (x - a) + func(a)

x = np.linspace(-3, 3)
y = quad_func(x)
plt.plot(x, y, label="y", linestyle="dashed")

# あらゆる点における接戦を描く
START_X = -3
END_X = 3
NUM_STEPS = 100
for i in range(NUM_STEPS):
    a = START_X + (END_X-START_X) * i / NUM_STEPS
    plt.plot(x, tangent(quad_func, x, a))

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.xlim([-3, 3])
plt.ylim([-10, 10])

plt.show()
```
