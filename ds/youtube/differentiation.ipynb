```python
import numpy as np
import matplotlib.pyplot as plt

# 二次関数を描画する
x = np.linspace(-3, 3)
y = x**2 - 1
plt.plot(x, y, linestyle="dashed") # 点線で描画

# xを0に近づける
x_lim = np.array([2, 1, 0.5, 0.1, 0.01, 0.001])
y_lim = x_lim**2 + 1  # y=x^2+1
plt.scatter(x_lim, y_lim)  # 散布図で表示

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

def my_func(x):  # y=x^2+1を表す関数
    y = x**2 + 1
    return y

x = np.linspace(-3, 3)
y = my_func(x)
plt.plot(x, y, label="y", linestyle="dashed")

## 注意！これは接線ぽくて接線じゃない！！
delta_x = 0.01  # 0に近い⊿xの値
delta_y = my_func(x+delta_x) - my_func(x)  # ⊿y
delta_y_delta_x = delta_y / delta_x  # ⊿x/⊿y
plt.plot(x, delta_y_delta_x, label="⊿y/⊿x")

plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

def my_func(x):  # y=x^2+1を表す関数
    y = x**2 + 1
    return y

def tangent(func, x, a):
    delta_x = 1e-4
    derivative = (func(a+delta_x) - func(a)) / delta_x
    return derivative * (x - a) + func(a)

x = np.linspace(-3, 3)
y = my_func(x)
plt.plot(x, y, label="y", linestyle="dashed")

## x = 4の時の接線
plt.plot(x, tangent(my_func, x, 2), label="⊿y/⊿x")

START_X = -2
END_X = 2
NUM_STEPS = 100
for i in range(NUM_STEPS):
    a = START_X + (END_X-START_X) * i / NUM_STEPS
    plt.plot(x, tangent(my_func, x, a), label="⊿y/⊿x")


# plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.xlim([-3, 3])
plt.ylim([-10, 10])

plt.show()
```
